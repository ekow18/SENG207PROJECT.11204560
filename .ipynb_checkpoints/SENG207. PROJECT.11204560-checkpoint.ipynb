```python
features = df.columns.drop('quality')
for feature in features:
   mplt.figure(figsize=(10, 6))
   mplt.scatter(df[feature], df['quality'])
   mplt.xlabel(feature, fontsize=14)
   mplt.ylabel('Quality', fontsize=14)
   mplt.title(f'Quality vs {feature}', fontsize=18)
   mplt.show()
```
